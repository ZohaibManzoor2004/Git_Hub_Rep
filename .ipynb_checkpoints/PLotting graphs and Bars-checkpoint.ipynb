## Adding titles at top of the sepal plot 

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

# draw a lineplot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("sepal plot")

plt.show()
```

## Adding Limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None, rc =None)
# load dataset
phool = sns.load_dataset("iris")
phool

# draw a lineplot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("sepal plot")
plt.xlim(2)
plt.ylim(2)

plt.show()
```

## Size of figure

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool
#Change of figure
plt.figure(figsize=(8,4))

# draw a lineplot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("sepal plot")

plt.show()
```

## Random y axis changing

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

# draw a lineplot
sns.barplot(x="species",y="petal_length",data=phool,order=["virginica","versicolor","setosa"])
# Here in order argument we've changed the order, the program can execute with or without it.


plt.show()    
```

## Arrangement of Order of Titanic data_set

## Color shift in data_set

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti
#Plottin
sns.barplot(x="sex",y="alone",hue="who",data=kashti,order=("female","male"),color=("red"))


plt.show()
```

## Removing Error Bars by using confindance intervals

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti
#Plottin
sns.barplot(x="sex",y="alone",hue="who",data=kashti,order=("female","male"),color=("red"),ci=None)


plt.show()
```

## Changing the color Paletts

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti
#Plottin
sns.barplot(x="sex",y="alone",hue="who",data=kashti,order=("female","male"),color=("red"),ci=None,palette="pastel")


plt.show()
```

## Applying  median:

```python
# import libraries
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti
#Plotting
sns.barplot(x="sex",y="alone",hue="who",data=kashti,estimator=median)


plt.show()
```

## Applying mean:

```python
# import libraries
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti
#Plotting
sns.barplot(x="sex",y="alone",hue="who",data=kashti,estimator=mean)


plt.show()
```

## Applying Saturation(Intensity of the colour of Bars)

```python
# import libraries
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti
#Plotting
sns.barplot(x="sex",y="alone",hue="who",data=kashti,estimator=mean,saturation=0.9)


plt.show()
```
